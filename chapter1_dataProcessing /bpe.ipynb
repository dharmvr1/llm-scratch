```python
class BpeTOkenizer:
    def __init__(self):
        self.vocab ={l: i for i,l in enumerate(string.ascii_letters)}
        self.merges =[]
    
    def training(self , text):
        self.vocab["</UNK>"] = len(self.vocab)
        split_text = re.findall(r"\w+|[^\w\s]", text)
        word_count  =  {}
        for i in split_text:
            if word_count.get(i):
                word_count[i][1]+=1
            else:
                word_count[i] =[list(i)+['</w>'] ,1]
        
        def _pair_count():
            pair_counting={}
            for key,values in word_count.items():
                # print(key)
                word_char =values[0]
                counting =values[1]
                
            
                for i in range(len(word_char)-1):
                    char1,char2 = word_char[i],word_char[i+1]
                    pair_counting[(char1,char2)]  =pair_counting.get((char1,char2) ,0) + counting
            
            return pair_counting
        pair_counting =_pair_count()
        best_next_pair  =max(pair_counting, key=pair_counting.get)
        n=1000
        while n>0:
            if len(pair_counting) ==0:
                break
            char1,char2 = best_next_pair
            
            self.merges.append((char1,char2))
            if self.vocab.get(char1) is None:
                # print(vocab)
                self.vocab[char1] = len(self.vocab)
            if  self.vocab.get(char2) is None:
                self.vocab[char2] = len(self.vocab)    
            # print(word_count)
            for word , values in word_count.items():
                
                word_char = values[0]
                i=0
                while i< len(word_char)-1:
                    if char1 ==word_char[i] and  char2 == word_char[i+1]: 
                        new_token = char1+char2
                        
                        if  self.vocab.get(new_token) is None:
                                self.vocab[new_token] = len(self.vocab) 
                        
                        word_char = word_char[:i] + [new_token]+ word_char[i+2:]
                        word_count[word][0] = word_char
                        
                        i+=2
                    else :
                        i+=1
                
            # word_count = pair_count(word_count)
            # print(word_count)
            pair_counting =_pair_count()
                # print(pair_counting)
            
                
            if len(pair_counting) ==0:
                break
                
            best_next_pair  =max(pair_counting, key=pair_counting.get)
            # print(best_next_pair)
            n-=1
    
    
    def encode(self,text):
        
        split_text = re.findall(r"\w+|[^\w\s]", text)
        
        encoding_text =[]
        
        for word_chars in split_text:
            word_chars = list(word_chars) +["</w>"]
            for char1 , char2 in self.merges:
                i=0
                while i  < len(word_chars)-1:
                    if char1 == word_chars[i] and char2 == word_chars[i+1]:
                        # print(word_chars)
                        word_chars = word_chars[:i] + [word_chars[i]+word_chars[i+1]] + word_chars[i+2:]
                        i+=2
                        
                    else :
                        if self.vocab.get(word_chars[i]) is None:
                            word_chars[i] = '</UNK>' 
                            # print(word_chars[i])
                        i+=1  
            encoding_text.append(word_chars)    
                
        return [self.vocab.get(i)  for word_encoding in encoding_text for i in word_encoding]
        
            
    def decode(self,token):
        deocde_vocab = {i:s for s,i in self.vocab.items()}
        
        return (''.join(deocde_vocab[i] for i in token)).replace('</w>', ' ')
    
        
        
        
```
